```python
import numpy as np
import matplotlib.pyplot as plt

a = 20
b = 0.2
c = 5

k = 0.5
h = 10
M_over_P = 20

Y = np.linspace(0, 200, 500)

r_is = (a - (1 - b) * Y) / c
r_lm = (k * Y - M_over_P) / h

A_shock = 5
r_is_shock = (A_shock - (1 - b) * Y) / c

idx_eq = np.argmin(np.abs(r_is - r_lm))
Y_eq = Y[idx_eq]
r_eq = r_is[idx_eq]

idx_eq_shock = np.argmin(np.abs(r_is_shock - r_lm))
Y_eq_shock = Y[idx_eq_shock]
r_eq_shock = r_is_shock[idx_eq_shock]

plt.plot(Y, r_is, label="IS")
plt.plot(Y, r_lm, label="LM")
plt.plot(Y, r_is_shock, label="IS after shock", linestyle="--")

plt.scatter(Y_eq, r_eq, label=f"Initial eq ({Y_eq:.1f}, {r_eq:.1f})")
plt.scatter(Y_eq_shock, r_eq_shock, label=f"Shock eq ({Y_eq_shock:.1f}, {r_eq_shock:.1f})")

plt.xlabel("Output (Y)")
plt.ylabel("Interest rate (r)")
plt.title("Simple IS-LM Model")
plt.legend()
plt.show()

print("Initial equilibrium:")
print("Y =", round(Y_eq, 2), "r =", round(r_eq, 2))

print("After negative demand shock:")
print("Y =", round(Y_eq_shock, 2), "r =", round(r_eq_shock, 2))
```
